O BlackJack (ou Vinte e Um) é um jogo em que os jogadores podem comprar cartas livremente, enquanto tiverem menos de 21 pontos. Ganha o jogador que tiver o maior número de pontos, desde que seja menor ou igual a 21. Nosso jogo deve ter as seguintes funções:

Função para o sorteio: essa função retira uma carta aleatória do baralho e retorna o número de pontos que essa carta vale.

Função verificação: verifica e indica qual/quais jogador/jogadores tem o maior número de pontos, que seja menor ou igual a 21.
